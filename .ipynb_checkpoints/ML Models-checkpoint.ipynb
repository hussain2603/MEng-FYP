```python
import pandas as pd
# import tensorflow as tf
# tf.compat.v1.logging.set_verbosity(tf.compat.v1.logging.ERROR)
# from keras.models import Sequential
# from keras.layers import Dense, Activation
# from keras.utils import np_utils
# from keras.datasets import cifar10
# import keras.optimizers 
# from keras.layers import Conv2D, MaxPooling2D, Flatten, AveragePooling2D


unstable_gen1 = pd.read_excel('Unstable_Points_Gen.xlsx', sheet_name='gen1')
stable_gen1  = pd.read_excel('Stable_Points_Gen.xlsx', sheet_name='gen1' )
```

```python
unstable_gen1
```

```python
stable_gen1
```

```python
points_gen1 = pd.concat([stable_gen1, unstable_gen1], ignore_index=True)

points_gen1
```

```python
points_gen1 = points_gen1.sample(frac=1).reset_index(drop=True)
points_gen1
```

```python
#split into training and test

points_gen1_train = points_gen1[:int(.8*len(points_gen1))]
points_gen1_test = points_gen1[int(.8*len(points_gen1)):]

points_gen1_test
```

```python
features = ['Rotor Angle', 'Active Power', 'Voltage Mag']

x_train = points_gen1_train[features]
y_train = points_gen1_train['stable']

x_test = points_gen1_test[features]
y_test = points_gen1_test['stable']

y_test
```

```python
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score, f1_score, mean_absolute_error, mean_squared_error, max_error

classifier = SVC(kernel='poly', degree=20)
classifier.fit(x_train, y_train)

y_pred = classifier.predict(x_test)
print(accuracy_score(y_test,y_pred))
print(f1_score(y_test, y_pred))
```
